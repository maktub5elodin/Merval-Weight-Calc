```python
import yfinance as yf
```

```python
import pandas as pd
```

```python
merv = ['ALUA.BA', 'BBAR.BA', 'BMA.BA', 'BYMA.BA', 'CEPU.BA', 'COME.BA', 'CRES.BA', 'EDN.BA', 'GGAL.BA', 'IRSA.BA', 'LOMA.BA', 'PAMP.BA', 'SUPV.BA', 'TECO2.BA', 'TGNO4.BA', 'TGSU2.BA', 'TRAN.BA', 'TXAR.BA', 'VALO.BA', 'YPFD.BA']

print(merv)

# Solo a fines de conseguir la lista de Tickers en Formato Yahoo Finance
```

```python
mervdf = yf.download(merv, period='1mo')
#mervdf = yf.download(merv, period='3mo')
#mervdf = yf.download(merv, period='6mo')
#mervdf = yf.download(merv, period='ytd')

# Descarga de datos para los ticker seleccionados --> cambiar el time lapse as disired
```

```python
mervdf2 = pd.DataFrame()

for i in merv:
    mervdf2[i] = (mervdf["Volume"][i] * mervdf["Close"][i] / 1000000)

# Crea un DataFrame (DF) vacío y lo rellena de info con el cálculo de volument efectivo (monto operado en Millones de ARS)
# tiene en cuento el precio al cierre --> Esto no es enteramente correcto pero sirve para trabajar data la info provista por YF
```

```python
mervdf2
```

```python
mervdf2["Effective_Volume"] = mervdf2.sum(axis = 1)

# Añade al DF la suma total del monto efectivo operado en el Mercado Argentino en una fecha determinada
```

```python
for i in merv:
    mervdf2[i] = (mervdf2[i] / mervdf2["Effective_Volume"] * 100)

# Modifica el DF con el porcentaje (%) q representa el ticker del monto operado en una fecha determinada
```
